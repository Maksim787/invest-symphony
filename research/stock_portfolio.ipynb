```python
from library.load import load_data

df = load_data()
```

```python
def plot_w(w: pd.Series):
    sns.barplot(x=w.index, y=w)
    plt.ylabel('%  ')
    plt.xticks(rotation=90)
```

```python
from library.markowitz import get_markowitz_w

mu_s = [10, 20, 30]
plt.subplots(1, 3, figsize=(20, 5))
w_s = [get_markowitz_w(df, mu_year_pct=mu) for mu in mu_s]
for i, (w, mu) in enumerate(zip(w_s, mu_s)):
    plt.subplot(1, 3, i + 1)
    plot_w(w * 100)
    plt.title(f'mu={mu}%')
```
